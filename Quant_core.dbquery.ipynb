```sql
--Dimension --
select * from quant_core.bronze.asset_classes_raw;
select * from quant_core.bronze.currencies_raw;
select * from quant_core.bronze.counterparties_raw;
select * from quant_core.bronze.instruments_raw;
select * from quant_core.bronze.portfolios_raw;
select * from quant_core.bronze.market_data_sources_raw;
--Fact --
select * from quant_core.bronze.cashflows_raw;
select cashflow_yyyymm,count(*) from quant_core.bronze.cashflows_raw 
group by all;
select * from quant_core.bronze.market_prices_daily_raw;
select source_yyyymm,count(*) from quant_core.bronze.market_prices_daily_raw 
group by all;
select * from quant_core.bronze.positions_daily_raw;
select source_yyyymm,count(*) from quant_core.bronze.positions_daily_raw 
group by all;
select * from quant_core.bronze.transactions_raw;

select source_yyyymm,count(*) from quant_core.bronze.transactions_raw 
group by all;




---Silver----

--Dimension --
select * from quant_core.silver.dim_asset_class;
select * from quant_core.silver.dim_currency;
select * from quant_core.silver.dim_counterparty;
select * from quant_core.silver.dim_instrument;
select * from quant_core.silver.dim_portfolio;
select * from quant_core.silver.dim_market_data_source;
--Fact --
select * from quant_core.silver.fact_cashflows;
select * from quant_core.silver.fact_market_prices_daily;
select * from quant_core.silver.fact_positions_daily;
select * from quant_core.silver.fact_transactions;


-- Verify both months are preserved after running the ETL
SELECT 'fact_transactions' as table_name, trade_yyyymm as event_yyyymm, COUNT(*) as record_count
FROM quant_core.silver.fact_transactions 
GROUP BY ALL

UNION ALL

SELECT 'fact_positions_daily', position_yyyymm, COUNT(*) 
FROM quant_core.silver.fact_positions_daily 
GROUP BY ALL

UNION ALL

SELECT 'fact_market_prices_daily', price_yyyymm, COUNT(*) 
FROM quant_core.silver.fact_market_prices_daily 
GROUP BY ALL

UNION ALL

SELECT 'fact_cashflows', cashflow_yyyymm, COUNT(*) 
FROM quant_core.silver.fact_cashflows 
GROUP BY ALL

ORDER BY table_name, event_yyyymm



```
